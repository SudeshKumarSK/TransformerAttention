# Positional Encodings - Capturing word order in Transformer Architecture
-----

### Transformer Deep Dive

## Positional Encodings - Overview

## Why we need Positional Encodings??

* The Transformer does not process the words in sequence rather it processed these words / tokens parallely. For each token, it combines information from other tokens using multi-headed self attention. We must input the order explicitly for the Transformer to be aware of the word order.

* Positional Encodings are kind of `identifiers` that give some signal to the Neural Network to make it understand the word order explicitly.

* In the semantic space of word embeddings, adding these positional encodings move them a little away in the positional directions giving indications to the Neural Net. This results in a little multi-dimensional shift.

## Requirements of the Positional Embeddings

1. Every position should have the same identifier irrespective of the `Sequence Lengths` or what exactly the input is.
2. They `should not be too large`, where they can push the embeddings too far where the positional similarity overtones the semantic similarity.
3. Due to these requirements, the research paper came up with a novel approach to use alternating sine and cosine signals to encode the positions in each dimension of the word embeddings.

### Why Sines and Cosines?
1. They are `bounded between -1 and 1` and also have high variability for longer sequences unline sigmoid which is also bounded but doesn't have variability.
2. We can also adjust the frequencies to get less repetitive values and also for each dimension of each position, we can use different frequencies to get different values.
3. We alternate between sines and cosines for different index to get variable values between two positions in the sequence.

### Important Reasons Why?
1. Periodicity - For eg: The third token in the input sequence during self-attention can attend to 5 words after it and 10 words after it followed by 15 words after it in a much more tractable way.
2. Contrained Values between -1 and 1. Not, pushing the tokens far away over-shadowing the semantics.
3. Easy to extrapolate for longer sequences.
